### Proccessing WikiText

```python
wiki = load_it('Dataset/WikiText/data/DATA(Wiki).pkl')
```

```python
regex = r"[(#)a-zA-Z]"
index = []
cleaned = []
for i, k in enumerate(wiki): 
    if re.search(regex, k) is not None: index.append(i)
```

```python
for i in index:
    test_str = wiki[i]
    subst = ""
    result = re.sub(regex, subst, test_str, 0, re.MULTILINE)
    if result: cleaned.append(result)
    else: print('Skipped')
```

```python
for i, k in zip(index, cleaned):
    wiki[i] = k
```
